```python
# 2
def remove_ith_char(string , char):
    if char > len(string) or char < 0:
        return "invalid index"

    else:
        new_string = string[:char] + string[char+1:]
    return new_string

string = input("enter your string : ")
char = int(input("enter the character you want to remove : "))

print("string after removing cahracter is : " , remove_ith_char(string , char))
```

```python
# 3
def split_and_join(string):
    splitted_string = string.split(" ")
    print(splitted_string)

    joined_string = "-".join(splitted_string)
    print(joined_string)
    return joined_string

string = input("enter the string : ")
result = split_and_join(string)
print("New string after split and joining : ",result)
```

```python
# 4
def binary_string(string):
    for i in string:
        if i == 1 or i == 0:
            return "string is binary"
        else:
            return "string is not binary"

string = input("enter the string : ")
binary_string(string)
```

```python
# 5
def uncommon_words(string1 , string2):
    list1 = string1.split()
    list2 = string2.split()
    
    uncommon_letters = []  

    for i in list1:
        if i not in list2:
            uncommon_letters.append(i)

    for i in list2:
        if i not in list1:
            uncommon_letters.append(i)
    
    return uncommon_letters
        
string1 = input("enter the first string :  ")
string2 = input("enter the second string :  ")

uncommon_words(string1 , string2)
```

```python
def find_uncommon_words(s1, s2):
    # Split strings into words
    words1 = set(s1.split())
    words2 = set(s2.split())

    # Find uncommon words using symmetric difference
    uncommon = words1.symmetric_difference(words2)

    return uncommon


# Taking inputs
string1 = input("Enter first string: ")
string2 = input("Enter second string: ")

result = find_uncommon_words(string1, string2)
print("Uncommon words are:", result)

```

```python
# 6
def duplicate_letters(string):
    word_list = []

    duplicate_letter = []
    for i in string:
        word_list.append(i)
    
    for char in word_list:
        char_count = word_list.count(char)
        if char_count > 1:
            duplicate_letter.append(char)

    return duplicate_letter    
     
string = input("enter the string : ")

duplicate_letters(string)
```

```python
# 7
def contains_special_char(s):
    for char in s:
        if not char.isalnum():   
            return True
    return False

text = input("Enter a string: ")

if contains_special_char(text):
    print("String contains special characters.")
else:
    print("String does NOT contain special characters.")
```
