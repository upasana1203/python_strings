```python
# 1
def return_list_without_strings(list):
    new_list = []
    for i in list:
        if type(i) == int:
            new_list.append(i)
        
    return new_list


list = ["a" , 0 ,"eDABIT" , 1729 , "PYTHON" ,"1729"]       
return_list_without_strings(list)
```

```python
# 2
def add_index_to_element(user_list):
    result = []
    for index , ch in enumerate(user_list):
        sum = index + ch
        result.append(sum)
    return result

input_list = [1,2,3,4,5]
add_index_to_element(input_list)
```

```python
# 3
import math

# for pi import math module instead of useing 3.14
def volume_of_cone(h, r):
    volume = (math.pi * r * r * h) / 3
    result = round(volume, 2)
    return result

print(volume_of_cone(15, 6))
```

```python
# 5
def missing_num(lst):
    for i in range(1, 11):      
        if i not in lst:         
            return i

list_input = [1,2,3,4,6,7,8,9,10]
missing_num(list_input)
```

```python
def missing_num(lst):
    full_set = set(range(1, 11))  # {1,2,...,10}
    lst_set = set(lst)
    missing = full_set - lst_set   # set difference gives missing number
    return missing.pop()           # convert set to single number
```

```python
def missing_num(lst):
    lst.sort()
    for i, num in enumerate(lst, 1):
        if i != num:
            return i
```
