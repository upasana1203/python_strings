```python
# 1
def next_in_line(lst, num):
    if len(lst) == 0:
        return "No list has been selected"
    
    lst.append(num)   
    lst.pop(0)      
    return lst

    
user_list = [5,6,7,8,9]

next_in_line(user_list , 5)
```

```python
# 2

def sum_of_budget(lst):
    sum = 0
    for i in lst:
        sum += i["budget"]
    return sum

liist = ([
{ "name": "John", "age": 21, "budget": 23000 },
{ "name": "Steve", "age": 32, "budget": 40000 },
{ "name": "Martin", "age": 16, "budget": 2700 }
])
        
sum_of_budget  (liist)  
```

```python
# 5
def return_list_without_strings(list):
    new_list = []
    for i in list:
        if type(i) == int:
            new_list.append(i)
        
    return new_list


list = ["hello" , 81 , "basketball" , 123 , "fox"]       
return_list_without_strings(list)
```
