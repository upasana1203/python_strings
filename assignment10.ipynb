```python
# 5
def second_largest(arr):
    largest = float('-inf')
    second = float('-inf')

    for num in arr:
        if num > largest:
            second = largest
            largest = num
        elif num > second and num != largest:
            second = num

    return second


arr = [10, 20, 4, 45, 99]
print(second_largest(arr))
```

```python
# 7
def even_num(nums):
    even = []
    for i in nums:
        if i % 2 == 0:
            even.append(i)
            
    return even




list_items = int(input("enter the number of elements in your array : "))
nums = []
for i in range(list_items):
    n = int(input(f"enter element {i} : "))
    nums.append(n)


print("even num in list are : ", even_num(nums))
```

```python
# 8
def odd_num(nums):
    odd = []
    for i in nums:
        if i % 2 != 0:
            odd.append(i)
            
    return odd




list_items = int(input("enter the number of elements in your array : "))
nums = []
for i in range(list_items):
    n = int(input(f"enter element {i} : "))
    nums.append(n)


print("odd num in list are : ", odd_num(nums))
```

```python
# 9
def remove_empty_list(nums):
    result = []
    for i in nums:
        if i != []:
            result.append(i)
    return result


list_items = int(input("Enter number of elements: "))
nums = []

for i in range(list_items):
    value = input(f"Enter element {i}: ")
    
    # Convert "[]" into actual empty list
    if value == "[]":
        nums.append([])
    else:
        nums.append(value)

print("After removing empty lists:", remove_empty_list(nums))
```
