```python
import pyttsx3

engine = pyttsx3.init()

def speak(text):
    engine.say(text)
    if not engine._inLoop:   # safety check
        engine.runAndWait() 

def birthday_dict(birthdays):
    speak("Enter the name of your friend")
    friend_name = input("Enter the name of your friend: ")
    

    if friend_name in birthdays:
        result = f"{friend_name}'s birthday is on {birthdays[friend_name]}"
        print(result)
        speak(result)

    elif friend_name not in birthdays:
        msg = f"I could not find {friend_name}'s birthday. Do you want to update it?"
        print(msg)
        speak(msg)

        speak("Enter Your Choice")
        choice = input("Your choice (yes/no): ")
        
        speak(f"your choice is {choice}")
    
        if choice.lower() == "yes":
            speak("enter your friend name to update")
            new_friend = input("Enter your new friend's name: ")
            
            speak("enter your friend bithday date to update")
            new_friend_bdate = input("Enter your friend's birthday date: ")
            
        
            birthdays[new_friend] = new_friend_bdate
            print(birthdays)
            speak(f"{new_friend}'s birthday has been updated.")
    
        elif choice.lower() == "no":
            speak("Thank you!")
        


    
birthdays = {'Alice': 'Apr 1', 'Bob': 'Dec 12', 'Carol': 'Mar 4', 'Sachin': 'Jul 6'}



birthday_dict(birthdays)
```
