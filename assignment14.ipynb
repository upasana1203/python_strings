```python
# 1
class Generator:
    def __init__(self , num):
        self.num = num

    def divisibleBySeven(self):
        for i in range(0 , self.num+1):
            if i % 7 == 0:
                yield i

num = int(input("enter the number : "))

obj = Generator(num)

for n in obj.divisibleBySeven():
    print(n)

                
```

```python
# 2
def word_count_in_sentence(sentence):
    words = sentence.split()
    counts = {}

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1
    
    sorted_words = sorted(counts.keys())

    for word in sorted_words:
        print(f"{word} : {counts[word]}")
            
            
inputt = input("enter the sentence : ")
word_count_in_sentence(inputt)
```

```python
# 4
def generate_sentence(subject , verb , objects):
    for i in subject:
        for j in verb:
            for k in objects:
                print(f"{i} {j} {k}")

subject = ["I" ,"You"]
verb = ["Play" , "Love"]
objects = ["Hockey" , "Football"]
generate_sentence(subject , verb , objects)
```

```python
# 6
def binary_search(arr , find):
    start = 0
    end = len(arr) - 1

    while start <= end:
        mid = (start + end) //2

        if arr[mid] == find:
            return f"the index of element is {mid}"
        elif arr[mid] < find:
            start = mid+1
        else:
            end = mid -1
            
    else:
        print("element is not available")

arr = [2, 4, 6, 8, 10, 12, 14]
find = int(input("enter the num from arr to find : "))

binary_search(arr , find)


    
```
