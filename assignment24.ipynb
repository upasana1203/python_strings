```python
# 1
def amplify(n):
    return [i * 10 if i % 4 == 0 else i for i in range(1, n + 1)]

amplify(12)
```

```python
# 2
def find_unique(lst): 
    for item in lst: 
        if lst.count(item)< 2: 
            print(f"{item} is unique element")

find_unique([3,3,3,7,3,3])
```

```python
# 3
from math import pi
class Circle:
    def __init__(self , r):
        self.r = r
    def get_area(self):
        area = 2 * pi*self.r*self.r
        return f"the area of circle is {area}"
    def get_circumference(self):
        circumference = 2 * pi *self.r
        return f"the circumference of circle is {circumference}"

first_circle = Circle(6)
print(first_circle.get_area())
print(first_circle.get_circumference())
```

```python
# 4
def sort_by_length(lst):
    for i in range(len(lst)):
        for j in range(i + 1, len(lst)):
            if len(lst[i]) > len(lst[j]):
                lst[i], lst[j] = lst[j], lst[i]
    return lst
sort_by_length(["google" , "apple" , "microsoft"])
sort_by_length(["leonardo" , "Michelangelo" , "raphel" , "donatello"])
```

```python
def sort_by_length(lst):
    return sorted(lst ,  key = len )

sort_by_length(["google" , "apple" , "microsoft"])
sort_by_length(["leonardo" , "Michelangelo" , "raphel" , "donatello"])
```

```python
# 5
def pythagoras(lst):
    sorted_num = sorted(lst)

    sqr_two = (sorted_num[0]**2 + sorted_num[1]**2)
    sqr_largest = sorted_num[2]**2

    if sqr_two == sqr_largest:
        return True
    else:
        return False
    

pythagoras([1,2,3])
```
