```python
# 1
def list_operation(x , y , n):
    divisible = []
    for i in range(x , y +1):
        if i % n == 0:
            divisible.append(i)
    return divisible

start = int(input("enter where to start : "))
end = int(input("enter where to end : "))
divisor = int(input("enter the number for divisibility : "))

list_operation(start , end , divisor)
```

```python
# 2
def simon_says(lst1, lst2):
    return lst1[:-1] == lst2[1:]

list1 = [1,2]
list2 = [5,1]
simon_says(list1 , list2)
```

```python
# 3
def society_name(lst):
    letters = []
    for name in lst:
        letters.append(name[0])   
    letters.sort()                
    return ''.join(letters)


society_name(["Adam" , "Sarah" , "Malcolm"])
        
        
```

```python
def society_name(lst):
    return ''.join(sorted([name[0] for name in lst]))

society_name(["Phoebe" , "Chandler" , "Rachel" , "Rose" , "Monica" , "Joey"])
```

```python
# 4
def is_isogram(word):
    word = word.lower()
    for char in word:
        if word.count(char) > 1:
            return False
    return True


is_isogram("consecutive")
```

```python
# 5
def is_in_order(s):
    return s == ''.join(sorted(s))

is_in_order("abc")
```
