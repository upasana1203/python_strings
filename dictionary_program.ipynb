```python
import pyttsx3

engine = pyttsx3.init()

def speak(text):
    engine.say(text)
    if not engine._inLoop:   # safety check
        engine.runAndWait()   # Run immediately

def product(product_dict):
    
    print(product_dict)
    speak("Enter the product name")
    user_input = input("Enter the product name: ")

    if user_input in product_dict:
        output = f"The price of {user_input} is {product_dict[user_input]}"
        print(output)
        speak(output)

    else:
        print("Product is not available")
        speak("Product is not available")
        print("Do you want to add any product?")
        speak("Do you want to add any product?")
        user_choice = input("Choice yes or no: ")
        speak(f"Your choice is {user_choice}")

        if user_choice == "yes":
            new_product_name = input("Enter the name of product to add: ")
            new_product_price = input("Enter the price of product: ")

            product_dict[new_product_name] = new_product_price

            print("Product added successfully")
            speak("Product added successfully")
            print("Updated list:", product_dict)

        elif user_choice == "no":
            print("User doesn't want to add any new product")
            speak("User doesn't want to add any new product")

        else:
            print("Exit")
            speak("Exit the app")

product_input = {
    "pen": 100,
    "notebook": 200,
    "highlighter": 80,
    "pencil": 20
}

product(product_input)
```
