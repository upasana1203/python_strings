```python
# 2

def add(*args):
    sum = 0
    for i in args:
        sum += i
    print(sum)

add(1,2,3,4,5,5,100)

def division(a , b):
    answer = a / b
    print(answer)
division(44,2)
    
```

```python
# 3
def triangle_area(breadth , height):
    area = 1/2*(breadth * height)
    print(area)
triangle_area(67 ,89)
```
