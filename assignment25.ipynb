```python
# 1

def find_count_of_similar(num1 , num2 , num3):
    if num1 == num2 and num1 == num3:
        return 3
    elif num1 == num2 or num1 == num3 or num2 == num3:
        return 2
    else :
        return 0


find_count_of_similar(1,1,1)
```

```python
# 2
def dict_to_list(d):
    return sorted(d.items())
dict_to_list({'d':1 , 'b' : 2 , 'c' : 6})
```

```python
# 3
def mapping(letters):
    result = {}
    for ch in letters:
        result[ch] = ch.upper()
    return result

mapping(["p" , "s" , "r" , "t"])
```

```python
# 4
def vow_replace(text, vowel):
    for v in "aeiou":
        text = text.replace(v, vowel)
    return text
vow_replace("apple and bananas" , "u")
vow_replace("stuffed jalapenno lapizoo" , "e")
```

```python
# 5
def ascii_capitalize(text):
    result = ""

    for ch in text:
        if ord(ch) % 2 == 0:
            result += ch.upper()
        else:
            result += ch.lower()

    return result
ascii_capitalize("once in a blue moon")
```
