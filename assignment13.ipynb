```python
# 1
import math 

def square_root(d , c , h):
    result = []
    for val in d:
        q = round(math.sqrt((2*c*val)/h) , 2)
        result.append(q)
    return result
    
c= 50
h = 30

d = [100,150,180]
square_root(d , c, h)
```

```python
# 2
def array_two_dime(x , y):
    arr = []

    for i in range(x):
        row = []
        for j in range(y):
            row.append(i*j)
        arr.append(row)
    return arr

x = int(input("enter value for row : "))
y = int(input("enter value for col : "))
array_two_dime(x , y)
```

```python
# 5
def count_letters_digits(sentence):
    # yes = sentence.isalnum()
    # if yes == True:
    #     print(sentence)
    # else:
    #     print("sentence is not alphanumeric")
    letter_count = 0 
    digit_count = 0
    for word in sentence:
        # print(word)
        letter_check = word.isalpha()
        digit_check = word.isnumeric()
        
        if letter_check == True:
            letter_count += 1
        elif digit_check == True:
            digit_count += 1
    print(f"LETTER = {letter_count}") 
    print(f"DIGIT = {digit_count}")
            
            
    
        

sentence = input("enter the sentence : ")
count_letters_digits(sentence)
```

```python
def check_valid_password(passwords):
    valid_pass = []

    for pwd in passwords:
        if len(pwd) < 6 or len(pwd) > 12:
            continue

        has_lower = False
        has_upper = False
        has_digit = False
        has_special = False

        for ch in pwd:
            if 'a' <= ch <= 'z':
                has_lower = True
            elif 'A' <= ch <= 'Z':
                has_upper = True
            elif '0' <= ch <= '9':
                has_digit = True
            elif ch in "$#@":
                has_special = True

        if has_lower and has_upper and has_digit and has_special:
            valid_pass.append(pwd)

    return valid_pass
                
        

# Taking input
passwords = input("Enter comma separated passwords: ").split(',')

# Checking valid ones
valid = check_valid_password(passwords)

# Printing result
print(",".join(valid))
```
