```python
# 2
def largest_number(arr):
    largest = arr[0]
    for i in arr:
        if i > largest:
            largest = i
    return largest

        
num = int(input("enter the numbers one by one : "))
arr = []

for i in range(num):
    n = int(input(f"Enter element {i+1}: "))
    arr.append(n)


largest_number(arr)
```

```python
# 3
def rotation(array , d):
    d = d % len(array)
    for _ in range(d):
        first = array.pop(0)
        array.append(first)
    return array

num = int(input("enter the numbers one after one : "))
array = []
for i in range(num):
    n = int(input(f"enter the element {i + 1} : "))
    array.append(n)

d = int(input("enter the number for rotations : "))

rotation(array , d)
    
```

```python
# 3
def rotation_right(arr , d):
    d = d % len(array)
    for _ in range(d):
        last = array.pop()
        array.insert(0 , last)
    return array

num = int(input("enter the numbers one after one : "))
array = []
for i in range(num):
    n = int(input(f"enter the element {i + 1} : "))
    array.append(n)

d = int(input("enter the number for rotations : "))

rotation_right(array , d)
    
```

```python
def monotonic(array):
    print("Array:", array)

    is_inc = True
    is_dec = True

    for i in range(1, len(array)):
        if array[i] < array[i-1]:
            is_inc = False
        if array[i] > array[i-1]:
            is_dec = False

    if is_inc or is_dec:
        print("Monotonic array")
    else:
        print("Not monotonic")


array_items = int(input("enter how many elements you want in array : "))
array = []
for i in range(array_items):
    n = int(input("enter the element {i} :"))
    array.append(n)

monotonic(array)
```
