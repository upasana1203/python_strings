```python

# 1
def factorial(num):
    factorial = 1
    for i in range(num,0,-1):
        factorial = factorial * i
        
    return factorial

print(f"the factorial of number is :",factorial(5))
        
```

```python
# 2
def multiplication_table(num):
    for i in range(1,11):
        print(f"{num} * {i} = {num * i} ")
multiplication_table(5)
```

```python
# 3
def fibonacci(n):
    a = 0
    b = 1
    
    # Check if n is less than 0
    if n < 0:
        print("Incorrect input")
        
    # Check if n is equal to 0
    elif n == 0:
        return 0
      
    # Check if n is equal to 1
    elif n == 1:
        return b
    else:
        for i in range(1, n):
            c = a + b
            print(f"{b} + {a} = {c}")
            a = b
            b = c
            print(b)
        return b

print(fibonacci(7))
    
```

```python
# 6
def sum(num):
    sum = 0
    if num <=0:
        return "invalid input"
    else:
        for i in range(num+1):
            sum += i
        return sum

sum(78)
```
