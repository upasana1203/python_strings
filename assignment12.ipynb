```python
def unique_values(dictionary):
    freq = {}
    
    for value in dictionary.values():
        freq[value] = freq.get(value, 0) + 1

    unique = [key for key, count in freq.items() if count == 1]
    return unique

print(unique_values(my_dict))
```

```python
from collections import Counter

def unique_values(dictionary):
    counter = Counter(dictionary.values())
    return [value for value, count in counter.items() if count == 1]

print(unique_values(my_dict))
```

```python
# 2

def sum_item(dictionary):
    sum = 0
    for value in dictionary.values():
        sum += value
    return sum
    
my_dict = {
    "one" : 111,
    "two" : 222,
    "three" : 333,
    "four" : 444
}

sum_item(my_dict)
```
