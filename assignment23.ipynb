```python
# 2
def multiply_nums(nums):
    nums = nums.replace(" ", "")      # remove any spaces
    numbers = nums.split(",")
    product = 1

    for n in numbers:
        product *= int(n)

    return product

print(multiply_nums("1,2, 3, 4"))
```

```python
# 3
def square_digits(num):
    result = ""

    for digit in str(num):        # iterate through each digit
        result += str(int(digit) ** 2)

    return int(result)
square_digits(9119)
```

```python
def square_digits(num):
    return int("".join(str(int(d)**2) for d in str(num)))

square_digits(9119)
```

```python
# 4
def setify(lst):
    remove_duplicate = set(lst)

    new_str= sorted(remove_duplicate)
    return new_str

setify([1,3,3,5,5])
    
```

```python
# 5

def mean(num):
    digits = str(num)               # convert number to string
    total = sum(int(d) for d in digits)   # sum of digits
    return total // len(digits)     # integer mean

mean(12345)
```
