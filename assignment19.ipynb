```python
# 1
def double_char(user_str):
    new_str = ""             
    for ch in user_str:
        new_str += ch * 2     
    return new_str

input_str = input("enter your string : ")
print(double_char(input_str))
```

```python
def double_char(s):
    return ''.join(ch*2 for ch in s)

input_str = input("enter your string : ")
print(double_char(input_str))
```

```python
# 2
def reverse_boolean(boolean_val):
    if boolean_val == True or boolean_val == False:
        return not boolean_val
    else:
        
        return "boolean expected"

input_bool = input("enter your value : ")

reverse_boolean(input_bool)
```

```python
def reverse_boolean(value):
    if isinstance(value, bool):      # checks if it's exactly a boolean
        return not value
    else:
        return "boolean expected"

input_bool = input("enter your value : ")

reverse_boolean(input_bool)
```

```python
# 4
def num_layers(n):
    thickness_mm = 0.5 * (2 ** n)  
    thickness_m = thickness_mm / 1000  
    return f"{thickness_m}m"

num_layers(6)
```

```python
# 4
def index_of_caps(user_str):
    index_Cap = []
    for position, ch in enumerate(user_str):  
        if ch.isupper():                 
            index_Cap.append(position)         
    return index_Cap

user_inp = input("enter the string: ")
print(index_of_caps(user_inp))
```

```python
# 5
def find_even(nums):
    return[i for i in range(nums+1) if i%2 == 0]

find_even(12)
```
