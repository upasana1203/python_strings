```python
def return_list_without_strings(items):
    new_list = []
    for i in items:
        if isinstance(i, int):
            new_list.append(i)
    return new_list


# Taking input from the user
user_input = input("Enter elements separated by spaces: ")

# Convert each element:
# If it's a digit → convert to int
# Otherwise → keep as string
items = []
for x in user_input.split():
    if x.isdigit():
        items.append(int(x))
    else:
        items.append(x)

# Passing to function
print(return_list_without_strings(items))
```

```python
# 2
def reverser(string):
    reversed_string = string[::-1].swapcase()
    # opp_case_reverse_str = reversed_string.swapcase()
    return reversed_string

str_name = input("enter the string : ")
reverser(str_name)
```

```python
def reverse(s):
    return "".join([ch.swapcase() for ch in s[::-1]])

reverser("Hello")
```

```python
# 3
# this code is right but does not use destructuring assignment(also called iterable unpacking)
def assign_item(user_list):
    first = user_list[0]
    middle = user_list[1:-1]
    last = user_list[-1]

    return first , middle , last
input_list = [1,2,3,4,5,6]
assign_item(input_list)
```

```python
# code with destructuring assignment
def assign_item(user_list):
    first, *middle, last = user_list
    return first, middle, last

input_list = [1, 2, 3, 4, 5, 6]
print(assign_item(input_list))
```

```python
# 4
def factorial(num):
    if num <= 1:
        return 1
    return num * factorial(num - 1)


user_num = int(input("enter the num :"))
factorial(user_num)
```

```python
# 5
def move_to_end(lst, element):
    result = []
    count = 0

    for x in lst:
        if x == element:
            count += 1
        else:
            result.append(x)

    result.extend([element] * count)
    return result

user_list = [1,3,2,4,4,1]
move_to_end(user_list,1)
```
